4. Make a table of all the producers and directors of each movie.

```python
#create a dataframe in which we have no null observations for the column crew
imdb_df_no_na_crew = imdb_df_new[~imdb_df_new['crew'].isna()]
```

```python
imdb_df_no_na_crew.shape
```

```python
imdb_df_no_na_crew.iloc[0]['crew']
```

```python
# Creating a function which returns the list of the producers and directors for each movie
def get_producers_and_directors(crew_list):
  try:
    #Initialize two lists corresponding to producer and director
    producer_list = []
    director_list = []

    for elem in crew_list :
      if elem['job'] == 'Producer':
        producer_list.append(elem['name'])

      if elem['job'] == 'Director':
        director_list.append(elem['name'])


    return producer_list,director_list
  except:
    return [],[]
```

```python
get_producers_and_directors(imdb_df_no_na_crew.loc[1,'crew'])
```

```python
imdb_df_new[['producers','directors']] = imdb_df_new.apply(lambda row:get_producers_and_directors(row['crew']),axis=1, result_type="expand")
```
